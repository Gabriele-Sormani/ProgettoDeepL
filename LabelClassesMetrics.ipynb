```python
test_imgs_path = '/content/gdrive/MyDrive/Progetto DeepL/test_imgs_2D.npy'
test_mask_path = '/content/gdrive/MyDrive/Progetto DeepL/test_labels_2D.npy'

X_test = np.load(test_imgs_path)
Y_test = np.load(test_mask_path)
```

```python
X_test = np.moveaxis(X_test, [1,3], [3,1])
Y_test = np.moveaxis(Y_test, [1,3], [3,1])
```

```python
Y_test.shape
```

# Prediction and metrics

```python
prediction = model.predict(X_test)
```

```python
prediction.shape
```

```python
iou_scores = {}
f1_scores = {}

iou_metric = IOUScore()
f1_metric = FScore()

num_classes = Y_test.shape[-1]
for label in range(num_classes):

    y_true_class = Y_test[..., label]
    y_pred_class = prediction[..., label]

    y_true_class_exp = np.expand_dims(y_true_class, axis=-1)
    y_pred_class_exp = np.expand_dims(y_pred_class, axis=-1)

    y_true_class_exp = tf.convert_to_tensor(y_true_class_exp, dtype=tf.float32)
    y_pred_class_exp = tf.convert_to_tensor(y_pred_class_exp, dtype=tf.float32)

    iou = iou_metric(y_true_class_exp, y_pred_class_exp).numpy()
    iou_scores[label] = iou

    f1 = f1_metric(y_true_class_exp, y_pred_class_exp).numpy()
    f1_scores[label] = f1


for label in range(num_classes):
    print(f"Class {label}: IoU = {iou_scores[label]:.4f}, F1 Score = {f1_scores[label]:.4f}")
```

```python
iou_tot = 0
for i in range(num_classes):
  iou_tot += iou_scores[i]

iou_tot = iou_tot/4

f1_tot = 0
for i in range(num_classes):
  f1_tot += f1_scores[i]

f1_tot = f1_tot/4
```
