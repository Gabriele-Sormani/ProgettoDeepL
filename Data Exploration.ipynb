# Label over image

```python
cervello_100 = '/content/destination_folder/Task01_BrainTumour/imagesTr/BRATS_100.nii.gz'
label_100 = '/content/destination_folder/Task01_BrainTumour/labelsTr/BRATS_100.nii.gz'
imgLabel = nib.load(label_100)
img = nib.load(cervello_100)
nii_data = img.get_fdata()
nii_data_labels = imgLabel.get_fdata()
```

```python
mid_slice_xLabel = nii_data_labels[:, :, 100]
mid_slice_x = nii_data[:, :, 100,2]
```

```python
mask = mid_slice_xLabel > 0
mid_slice_xLabel_alpha = np.zeros_like(mid_slice_xLabel)
mid_slice_xLabel_alpha[mask] = 1
```

```python
plt.imshow(mid_slice_x.T, cmap='gray', origin='lower')
plt.imshow(mid_slice_xLabel.T, cmap='YlOrRd', origin='lower', alpha=mid_slice_xLabel_alpha.T)
plt.xlabel('First axis')
plt.ylabel('Second axis')
plt.colorbar(label='Signal intensity')
plt.show()
```

* **1** (arancione chiaro): edema
* **2** (arancione scuro) : non-enhancing tumor
* **3** (rosso) : enhancing tumor
